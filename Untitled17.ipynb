```python
import re
import hashlib
from datetime import datetime
```

```python
selected_data =[]

def is_validn(phoneNumber):
    phone_num = re.sub(r'\D',"",phoneNumber)
    
    
    if re.match(r'^(\+?91|0?91)?[6-9]\d{9}$',phoneNumber):
        return True
    else:
        return False
    
def hashi(phoneNumber):
    return hashlib.sha256(phoneNumber.encode()).hexdigest()
    
def calculate_age(dob):
    if dob:
        dob = datetime.strptime(dob, '%Y-%m-%dT%H:%M:%S.%fZ')
        today = datetime.today()
        age = today.year - dob.year
        if today.month < dob.month or (today.month == dob.month and today.day < dob.day):
            age -= 1
        return age
    else:
        return None
    
for item in data:
    
    gender = item['patientDetails'].get('gender')
    fullName = item['patientDetails']['firstName'] + " " +  item['patientDetails']['lastName']
    phone_number = item['phoneNumber']
    is_valid = is_validn(phone_number)
    dob = item['patientDetails'].get('birthDate',"")
    age = calculate_age(dob)
    appointment_id = item['appointmentId']
    no_of_medicines = 0
    no_of_active_medicines = 0
    no_of_inactive_medicines = 0
    


    consultation_data = item.get('consultationData', {})
    medicines = consultation_data.get('medicines', [])

    for medicine in medicines:
        is_active = medicine.get('IsActive', False)
        no_of_medicines += 1
        if is_active:
            no_of_active_medicines += 1
        else:
            no_of_inactive_medicines += 1
    
    
    
    if is_valid:
        phone_num_hash = hashi(phone_number)
    else:
        phone_num_hash = None
        



    if gender=='M':
        gender='male'
    elif gender=='F':
        gender='female'
    else:
        gender = "null"
        
    #Error Implementing the names of medicines
    
    #medicines = item['consultationData'].get('medicines',[])
    #active_medicines = [medicine['name'] for medicine in medicines if medicine('IsActive',False)]
    
    selected_item = {
        "appointmentId":item['appointmentId'],
        "firstName":item['patientDetails']['firstName'],
        "lastName":item['patientDetails']['lastName'],
        "gender":gender,
        "DOB":item['patientDetails'].get('birthDate',""),
        "medicines":item['consultationData'].get('medicine',""),
        "fullName":fullName,
        "isValidMobile":is_valid,
        "phoneNumberHash":phone_num_hash,
        "Age":age,
        "no_of_medicines":no_of_medicines,
        
        "no_of_active_medicines":no_of_active_medicines,
        "phonNumber":phone_number,
        
        "no_of_inactive_medicines":no_of_inactive_medicines,
        #"medicineNames":", ".join(active_medicines)


        
    }
    
    
selected_data.append(selected_item)
```

```python
#plotting

import matplotlib.pyplot as plt

gender_counts = {'male': 0, 'female': 0, 'null': 0}

for item in selected_data:
    gender = item.get('gender', 'Other')
    gender_counts[gender] += 1

genders = list(gender_counts.keys())
appointments = list(gender_counts.values())

plt.pie(appointments, labels=genders, autopct='%1.1f%%')
plt.title('Number of Appointments by Gender')

plt.show()
```

```python
#converting to json
import json
from collections import defaultdict

aggregated_data = defaultdict(int)
valid_phone_numbers = set()

for item in selected_data:
    age = item.get('Age')
    gender = item.get('gender')
    isValidMobile = item.get('isValidMobile', False)
    no_of_medicines = item.get('noOfMedicines', 0)
    no_of_active_medicines = item.get('noOfActiveMedicines', 0)

    if age:
        aggregated_data['Age'] += age
    if gender:
        aggregated_data['gender'] += 1
    if isValidMobile:
        valid_phone_numbers.add(item['phoneNumber'])
        aggregated_data['validPhoneNumbers'] += 1
    aggregated_data['appointments'] += 1
    aggregated_data['medicines'] += no_of_medicines
    aggregated_data['activeMedicines'] += no_of_active_medicines

json_data = {
    'Age': aggregated_data['Age'],
    'gender': aggregated_data['gender'],
    'validPhoneNumbers': aggregated_data['validPhoneNumbers'],
    'appointments': aggregated_data['appointments'],
    'medicines': aggregated_data['medicines'],
    'activeMedicines': aggregated_data['activeMedicines']
}

#add ur path
with open('=aggregated_data.json', 'w') as file:
    json.dump(json_data, file)
```
